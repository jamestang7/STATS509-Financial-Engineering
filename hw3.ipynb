## (a)

$\mathbb{E}(f_b(x)) =\mathbb{E}(\frac{1}{n}\sum_{n=1}^{n}K_b(X-X_i))$

K_b(X-X_i) is iid to some pdf

$= \mathbb{E}(K_b(X-X_i))$

$=\int_{x = x - w/2}^{x+w/2}K_b(x-x')f(x')dx'$

$= \frac{1}{w}(F(x+w/2)-F(x-w/2))$, where w = 3.464b

$= \frac{1}{3.464b}(F(x+1.732b)-F(x-1.732b))$

## (b)

Bias = $\frac{1}{3.464b}(F(x+1.732b)-F(x-1.732b)) - f_X(x)$

```r
Bias_func = function(x,b){
    bias = 1/(3.464 * b)*(pged(q = x + 1.732*b, mean = 0,sd = 1, nu = 1.5) - 
                          pged(q = x - 1.732*b, mean = 0,sd = 1, nu = 1.5))-
                        dged(x = x, mean = 0,sd = 1, nu = 1.5)
    return(bias)
}
par(mfrow = c(1,3))
for (b in c(.2,.4,.6)){
    x = seq(-10,10,0.1)
    bias = Bias_func(x, b)
    plot(x = x, y = bias,xlab = "x",ylab = "Bias",main = paste("Bias when b=",b),ylim = c(-0.1,0.1))
}
```

## (c) 
$Var(f_b(x)) = \frac{1}{n}Var(K_b(X-X_i))$

$=\frac{1}{n}[\mathbb{E}(K_b^2(X-X_i))-\mathbb{E}^2(K_b(X-X_i)]$

$ = \frac{1}{n}\{\frac{1}{3.464b}^2[F(x+1.732b)-F(x-1.732b)] - [\frac{1}{3.464b}(F(x+1.732b)-F(x-1.732b))]^2 \}$

$ = \frac{1}{n}\frac{1}{3.464b}^2[F(x+1.732b)-F(x-1.732b)][1-F(x+1.732b)+F(x-1.732b))]$

$sd = \sqrt{Var(f_b(x))}$

```r
# note that var doesn't include n term
Var_func = function(x,b){
    var = 1/(3.464 * b)^2 *(pged(q = x + 1.732*b, mean = 0,sd = 1, nu = 1.5) - 
                          pged(q = x - 1.732*b, mean = 0,sd = 1, nu = 1.5))*
                        (1-pged(q = x + 1.732*b, mean = 0,sd = 1, nu = 1.5)+pged(q = x - 1.732*b, mean = 0,sd = 1, nu = 1.5))
    return(var)
}
par(mfrow = c(1,3))
for (b in c(.2,.4,.6)){
    x = seq(-10,10,0.1)
    var = Var_func(x, b)/sqrt(200)
    plot(x = x, y = var,xlab = "x",ylab = "SD",main = paste("Var when b=",b),ylim = c(-0.1,0.1))
}
```

## (d)

```r
par(mfrow = c(1,3))
df = data.frame(x=seq(-10,10,0.1))
for (b in c(.2,.4,.6)){
    x = seq(-10,10,0.1)
    mse = Var_func(x, b)/200 + Bias_func(x, b)^2
    plot(x = x, y = mse,xlab = "x",ylab = "MSE",main = paste("MSE when b=",b),ylim = c(-0.1,0.1),)
}
```

```r
df$a = Var_func(df$x, 0.2)/200 + Bias_func(df$x, 0.2)^2
df$b = Var_func(df$x, 0.4)/200 + Bias_func(df$x, 0.4)^2
df$c = Var_func(df$x, 0.6)/200 + Bias_func(df$x, 0.6)^2
head(df)
```

```r
t = ggplot( data = df, aes(x = x)) + 
      geom_line(aes(y = a, colour = "0.2")) +
      geom_line(aes(y = b, colour = "0.4")) +
    geom_line(aes(y = c, colour = "0.6")) +
    xlim(-5,5)+
    labs(x = "X", y = "MSE",title ="MSE with different b")+
    scale_colour_manual(name = "b value", values = c("0.2" = "red", "0.4" = "blue","0.6" = "green"))
t 
```

Based on the MSE graph, we should prefer 0.4
